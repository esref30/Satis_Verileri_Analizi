```python
data = df.drop(["Row ID","Order ID","Customer ID","Product ID"], axis = 1)
data.head()
```

```python
data["Postal Code"]= data["Postal Code"].fillna("5402")
```

```python
data.isnull().sum()
```

```python
data[data["City"]=="Burlington"]
```

```python
#ay stunu oluturma
data["Month"] = data["Order Date"].str[3:5].astype(int)
data
```

```python
#aylar isimlere dntrme
d = {1:"Ocak",2:"ubat",3:"Mart",4:"Nisan",5:"Mays",6:"Haziran",7:"Temmuz",8:"Austos",9:"Eyll",10:"Ekim",11:"Kasm",12:"Aralk"}
d
```

```python
data["Month"] = [d[i] for i in data["Month"]]
data
```

```python
data["Month"].value_counts()
```

```python
#hangi ayda sat fazla
sales_month = data.groupby("Month").agg({"Sales":sum})
sales_month.sort_values("Sales",ascending = False)
```

```python
#daha dzenli bir analiz yaplabilmesi iin aylar stun verisini index olmaktan kardm
sales_month.reset_index(inplace = True)
sales_month
```

```python
#aylara gre analiz ettiim sat verisini grselletirdim
plt.bar(sales_month["Month"],sales_month["Sales"])
plt.xlabel("Aylar")
plt.ylabel("Sales")
plt.xticks(fontsize = 14,rotation = 90)
plt.yticks(fontsize = 14)
plt.show()
```

```python
#zaman serisi analizini rahat yapabilmek iin datetime formatna evirdim
data["Order Date"] = pd.to_datetime(data["Order Date"],format='%d/%m/%Y')
data.head(10)
```

```python
data["Year"] = data["Order Date"].dt.year
data.head()
```

```python
year_sales = data.groupby(["Year"]).agg({"Sales":sum}).reset_index()
year_sales
```

```python
#YILLARA GRE SATI MKTARI
plt.bar(year_sales["Year"],year_sales["Sales"])
plt.xlabel("Yllar")
plt.ylabel("Toplam Sat")
plt.xticks(fontsize = 12,rotation = 90)
plt.yticks(fontsize = 12)
plt.show()
```

```python
#EN FAZLA SATI GEREKLETRLEN LK 20 MTER
customer_sales = data.groupby("Customer Name").agg({"Sales":sum}).reset_index()
customer_sales.head(20)
```

```python
customer_head20 = customer_sales.head(20)
plt.bar(customer_head20 ["Customer Name"],customer_head20 ["Sales"])
plt.xlabel("Mteriler")
plt.ylabel("Sat Miktar")
plt.xticks(fontsize = 12,rotation = 90)
plt.yticks(fontsize = 12)
plt.show()
```

```python
category_sales = data.groupby("Category").agg({"Sales":sum}).reset_index()
category_sales.head()
```

```python
#Satlan rnlerin Analizi
plt.pie(category_sales["Sales"],labels = category_sales["Category"],autopct = "%0.2f%%",explode = [0.1,0.1,0.1])
```
